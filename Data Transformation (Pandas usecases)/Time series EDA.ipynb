#### Time Resampling
https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78

```python
# Rule A : Year end frequency
df_tesla.resample(rule='A').min() 
# Provides information regarding min value for High, Low, Open, Close, Volume and Adj Close for respective years
```

```python
df_tesla.resample(rule='A').max()['Open'].plot();
```

```python
## Quarterly start frequency
df_tesla.resample(rule='QS').max()['Open'].plot(figsize=(10,4))
```

```python
# Business End frequency
df_tesla.resample(rule='BA').max()
```

```python
# Monthly
df_tesla['Open'].resample(rule='M').mean().plot(kind='bar', figsize=(15,6))
```

```python
# Business Quarter Start
df_tesla.resample(rule='BQS').max()
```

#### Rolling : Provides rolling window calculation
Grabbing n number of rows and applying aggregate function
Can also be considered as smoothening zig-zag graph

```python
df_tesla['High'].rolling(10).mean() 
# Grabbing 10 rows at a time and applying mean 
```

```python
df_tesla['Open: 30 days rolling'] = df_tesla['Open'].rolling(30).mean()
df_tesla[['Open', 'Open: 30 days rolling']].plot(figsize=(15,8));
```

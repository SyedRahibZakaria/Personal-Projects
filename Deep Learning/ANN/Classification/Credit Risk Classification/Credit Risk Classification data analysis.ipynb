```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('customer_data.csv')
```

```python
df.head()
```

```python
# Extracting feasture having missing values
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
# filling missing values with median as median is not affected by outliers
df['fea_2']=df['fea_2'].fillna(df['fea_2'].median())
```

```python
df['fea_2'].isna().sum()
```

```python
# Finding correlation between features
corr = df.corr()
plt.figure(figsize=(8,8))
sns.heatmap(corr, annot=True, cmap='Blues');
```

There isin't much information regarding what features we are working on
There is significant positive correlation (although not strong correlation) between fea_1 and fea_6, fea_3 and fea_7 and fea_10 and fea_11. 
There are some features with negative correlation which we might want to investigate further

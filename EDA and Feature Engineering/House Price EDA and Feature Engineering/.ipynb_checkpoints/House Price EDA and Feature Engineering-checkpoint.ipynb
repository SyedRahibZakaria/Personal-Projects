```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
pd.pandas.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('dataset.csv')
```

```python
print(df.shape)
```

```python
df.head()
```

In data analysis we will analyze to find out the following:
    1. Missing Values
    2. All the Numerical Variables
    3. Distribution of the Numerical Variables
    4. Cardinality of Categorical Variables
    5. Outliers
    7. Relationship between independent and dependent feature (SalePrice)

#### Missing Values

```python
# Grabbing the features having null values
features_with_na = [features for features in df.columns if df[features].isna().sum()>0]
```

```python
# Printing feature names along with percentage of missing values
for features in features_with_na:
    print(features, np.round(df[features].isna().mean(),4), '% missing values')
```

```python
# Plotting the features and checking if the variable has importance with SalesPrice

for feature in features_with_na:
    data = df.copy()
    
    # record that had nan value is converted to 1 and non-nan value to 0
    # to create count plot
    data[feature] = np.where(data[feature].isnull(), 1, 0) 
    
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

Some features has more impact on SalesPrice and plays and important role

```python
# Dropping id column as it is not important to us
df.drop('Id',axis=1, inplace=True)
df.head()
```

#### Extracting Numerical Variables

```python
num_features = [feature for feature in df.columns if df[feature].dtypes != 'O']
```

```python
df[num_features].head()
```

```python
# Plotting median SalesPrice vs Year Sold
df.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Price')
plt.title('Year Sold vs Price')
```

```python
# Grabbing feature related to year feature
yr_features = [feature for feature in df.columns if 'Yr' in feature or 'Year' in feature]
yr_features
```

```python
for feature in yr_features:
    if feature != 'YrSold':
        data = df.copy()
        # Difference in years
        data[feature] = data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

As the property gets old the SalePrice decreases

```python
# Numerical values are of two types: continuous and discrete

dis_features=[feature for feature in num_features if len(df[feature].unique())<25 and feature not in yr_features]
print('Total discrete features:',len(dis_features))
```

```python
df[dis_features].head()
```

```python
# Finding relationship between discrete features and salesPrice

for feature in dis_features:
    
    data=df.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(f'{feature} vs SalePrice')
    plt.show()
```

```python
# Grabbing continuous variables
cont_var = [feature for feature in num_features  if feature not in dis_features+yr_features]
print(f'Total numer of continuous variable columns: {len(cont_var)}')
```

```python
for feature in cont_var:
    data=df.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.show()
```

Note: Whenever we are trying to solve a regression problem, we convert non-gaussian distribution into gaussian distribution or standard normal distribution using logarithmic transformation

```python
for feature in cont_var:
    data=df.copy()
    if 0 in data[feature].unique(): # log0 is undefined
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(f'{feature} vs SalePrice')
        plt.show()
```

GrLivArea and 1stFlrSF has positive correlation (monotonic relationship) with SalePrice

#### Outliers

```python
# for continuous variable
for feature in cont_var:
    data=df.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.title(feature)
        plt.show()
```

#### Working with Categorical Variables

```python
cat_variables=[feature for feature in df.columns if df[feature].dtypes == 'O']
df[cat_variables].head()
```

```python
# Different categories for all the categorical variables
for feature in cat_variables:
    print(f'{feature} has {df[feature].nunique()} unique values')
```

```python
# Finding relationship between categorical variables and SalePrice

for feature in cat_variables:
    data=df.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.show()
```

#### Feature Engineering

```python
# Handling missing values for categorical variables
cat_feature_nan = [feature for feature in df.columns if df[feature].isna().sum()>0 and df[feature].dtypes=='O']

# Percentage of missing values for respective categorical variables

for feature in cat_feature_nan:
    print(f'{feature} has {np.round(df[feature].isna().mean(),4)} % missing values')
```

```python
# Replacing missing values with mode
for feature in cat_feature_nan:
    df[feature].fillna(df[feature].mode()[0],inplace=True)
df[cat_feature_nan].isna().sum()
```

```python
# grabbing numerical variables and checking for missing values
num_variable=[feature for feature in df.columns if df[feature].isna().sum()>0 and df[feature].dtypes != 'O']
```

```python
for feature in num_variable:
    print(f'{feature} has {np.round(df[feature].isna().mean(),4)} % missing values')
```

```python
# Replacing missing values with median

for feature in num_variable:
    median_value = df[feature].median()
    df[feature].fillna(median_value,inplace=True)
df[num_variable].isna().sum()
```
